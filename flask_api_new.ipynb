```python
import pandas as pd
from flask import Flask, jsonify, request
import pickle
import numpy
import flasgger
from flasgger import Swagger

# app
app = Flask(__name__)
Swagger(app)

# load model
pickle_in = open('model.pkl','rb')
model = pickle.load(pickle_in)

pickle_in = open('model1.pkl','rb')
model1 = pickle.load(pickle_in)

@app.route('/')
def welcome():
    return 'Welcome All'
     
# routes

@app.route('/predict_file' ,methods = ['POST'])
def predict_file():
    
    """Let's Predict the response 
    This is using docstrings for specifications.
    ---
    parameters:
      - name: file
        in: formData
        type: file
        required: true
      
    responses:
        200:
            description: The output values
        
    """
    
    
    df_test = pd.read_csv(request.files.get('file'))
    prediction = model.predict(df_test)
    return "The respone is" +str(list(prediction))


@app.route('/predict' ,methods = ['GET'])
def predict_response():
    """"Let's Predict the response
    This is using docstrings for specifications.
    ---
    parameters:  
      - name: Driving_License
        in: query
        type: number
        required: true
      - name: Region_Code
        in: query
        type: number
        required: true
      - name: Previously_Insured
        in: query
        type: number
        required: true
      - name: Vehicle_Age
        in: query
        type: number
        required: true
      - name: Annual_Premium
        in: query
        type: number
        required: true
      - name: Policy_Sales_Channel
        in: query
        type: number
        required: true
      - name: Vintage
        in: query
        type: number
        required: true
      - name: Gender_Male
        in: query
        type: number
        required: true
      - name: Vehicle_Damage_Yes
        in: query
        type: number
        required: true
    
    responses:
        200:
            description: The output values
        
    """
    
    Driving_License =request.args.get('Driving_License')
    Region_Code = request.args.get('Region_Code')
    Previously_Insured = request.args.get('Previously_Insured')
    Vehicle_Age = request.args.get('Vehicle_Age')
    Annual_Premium = request.args.get('Annual_Premium')
    Policy_Sales_Channel = request.args.get('Policy_Sales_Channel')
    Vintage = request.args.get('Vintage')
    Gender_Male = request.args.get ('Gender_Male')
    Vehicle_Damage_Yes = request.args.get('Vehicle_Damage_Yes')
    prediction = model1.predict([[Driving_License,Region_Code,Previously_Insured,Vehicle_Age,Annual_Premium,Policy_Sales_Channel,Vintage,Gender_Male,Vehicle_Damage_Yes]])
    return "The Response is" +str(prediction)





if __name__ == '__main__':
    app.run()
```
